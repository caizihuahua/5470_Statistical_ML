```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plsdata = pd.read_csv("./pls.csv")
```

```python
plt.figure(figsize=(5,3))
plt.title("PLS")
plt.xlabel('Year')
plt.ylabel('$R^2$')
plt.bar(np.arange(1988,2017),plsdata['r^2'][1:])
plt.tight_layout()
```

```python

plt.figure(figsize=(5,3))
plt.title("PLS")
plt.xlabel('Year')
plt.ylabel('# of Comp.')
plt.plot(np.arange(1987,2017),plsdata['K'])
plt.tight_layout()
```
